```python
df['Med_120'] = df.Close.rolling(120).mean()
df['Med_60'] = df.Close.rolling(60).mean()
df['Med_20'] = df.Close.rolling(20).mean()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def crave_column(lista_de_valores):
  inicio = lista_de_valores[0]
  fim = lista_de_valores[-1]
  meio = lista_de_valores[round(len(lista_de_valores)/2)]
  segundo = lista_de_valores[round(len(lista_de_valores)/100*25)]
  terceiro = lista_de_valores[round(len(lista_de_valores)/100*75)]
  med_inicio = np.mean([inicio,segundo])
  med_fim = np.mean([terceiro,fim])
  #plt.plot([inicio,segundo,meio,terceiro,fim])
  #plt.show()
  #plt.plot(lista_de_valores)
  #plt.show()
  veredito = 'acima' if med_fim > med_inicio and med_fim > meio else 'abaixo' if med_inicio > med_fim and med_inicio > meio else 'nivel'
  return veredito

close_temp = []
for idx, row in df.iterrows():
  close_temp.append(row.Close)
  if len(close_temp) >= 120:
    # 120
    df.loc[idx, '120_'] = crave_column(close_temp)
    # 60
    df.loc[idx, '60_'] = crave_column(close_temp[60:])
    # 20
    df.loc[idx, '20_'] = crave_column(close_temp[100:])
    #
    close_temp.pop(0)

df
```

```python
df.dropna(inplace=True)

dum_120 = pd.get_dummies(df['120_'], prefix='120')
dum_60 = pd.get_dummies(df['60_'], prefix='60')
dum_20 = pd.get_dummies(df['20_'], prefix='20')

df.drop(columns=['120_','60_','20_'], inplace=True)

df
```
